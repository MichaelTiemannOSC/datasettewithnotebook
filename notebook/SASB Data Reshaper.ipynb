```python
df = pd.read_excel("/home/jovyan/work/data/SASB dummy data set for testing UI 5-11-21 no formulas.xlsx",skiprows=1)
df
```

```python
id_columns=df.columns[4:19]
value_columns=df.columns[19:]
display(id_columns)
display(value_columns)
```

```python
df_melted=pd.melt(df, id_vars=id_columns, value_vars=value_columns).dropna(subset=['Sub-Metric Code', 'value'])
df_coded=df_melted[['Sub-Metric Code OR Sub-Activity Metric Code', 'Activity Metric Code', 'Sector Code', 'Industry Code', 'Disclosure Topic Code', 'Accounting Metric Code', 'Accounting Metric Category', 'Accounting Metric Unit of Measure', 'Sub-Metric Code', 'Sub-Metric Unit of Measure', 'variable', 'value']]
display(df_coded)
```

```python
df_coded.to_csv('/home/jovyan/work/data/SASB-coded.csv')
```

```python
sasb_sector = df_melted.groupby('Sector Code')['Sector'].agg('unique').apply(lambda x: x[0]).reset_index()
display(sasb_sector)
```

```python
sasb_industry = df_melted.groupby('Industry Code')['Industry'].agg('unique').apply(lambda x: x[0]).reset_index()
display(sasb_industry)
```

```python
sasb_disclosure_topic = df_melted.groupby('Disclosure Topic Code')['Disclosure Topic'].agg('unique').apply(lambda x: x[0]).reset_index()
display(sasb_disclosure_topic)
```

```python
sasb_accounting_metric = df_melted.groupby('Accounting Metric Code')['Accounting Metric or Activity Metric'].agg('unique').apply(lambda x: x[0]).reset_index()
display(sasb_accounting_metric)
```

```python
sasb_sub_metric = df_melted.groupby('Sub-Metric Code')['Sub-Metric OR Sub-Activity Metric'].agg('unique').apply(lambda x: x[0]).reset_index()
display(sasb_sub_metric)
```
